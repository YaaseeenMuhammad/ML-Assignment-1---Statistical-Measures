## Detecting the outliers using following methods and remove it using methods like trimming / capping/ imputation using mean or median

### a) Mean and Standard deviation

```python
mean = df['price_per_sqft'].mean()
median = df['price_per_sqft'].median()
std = df['price_per_sqft'].std()
MinimumBound = mean - 3 * std
MaximumBound = mean + 3 * std
outliers = df[(df['price_per_sqft'] < MinimumBound) | (df['price_per_sqft'] > MaximumBound)]
non_outliers = df[(df['price_per_sqft'] > MinimumBound) & (df['price_per_sqft'] < MaximumBound)]

print('\nMean\n', '_'* 25, '\n')
print(mean)
print('\n','-' * 25,'\n')
print('\nMedian\n', '_'* 25, '\n')
print(median)
print('\n','-' * 25,'\n')
print('\nStandard Deviation\n', '_'* 25, '\n')
print(std)
print('\n','-' * 25,'\n')
print('\nMinimum Bound\n', '_'* 25, '\n')
print(MinimumBound)
print('\n','-' * 25,'\n')
print('\nMaximum Bound\n', '_'* 25, '\n')
print(MaximumBound)
print('\n','-' * 25,'\n')
print('\nOUTLIERS\n', '_'* 25, '\n')
print(outliers)
print('\n','-' * 25,'\n')
print('\nNON OUTLIERS\n' + '_' * 25)
print(non_outliers)
```

### b) Percentile method

```python
lower_limit=df['price_per_sqft'].quantile(0.05)
upper_limit=df['price_per_sqft'].quantile(0.95)

p_outliers=df[(df['price_per_sqft']<=lower_limit) | (df['price_per_sqft']>=upper_limit)]
p_non_outliers=df[(df['price_per_sqft'] >=lower_limit) & (df['price_per_sqft'] <=upper_limit)]

print('\nLower Limit\n', '_'* 25, '\n')
print(lower_limit)
print('\n','-' * 25,'\n')
print('\nUpper Limit\n', '_'* 25, '\n')
print(upper_limit)
print('\n','-' * 25,'\n')
print('\nOUTLIERS\n', '_'* 25, '\n')
print(p_outliers)
print('\n','-' * 25,'\n')
print('\nNON OUTLIERS\n' + '_' * 25)
print(p_non_outliers)
```

### c) IQR(Inter quartile range method)

```python
Qtr1=df['price_per_sqft'].quantile(0.25)
Qtr3=df['price_per_sqft'].quantile(0.75)

IQR=Qtr3-Qtr1
LowerBound=Qtr1-1.5*IQR
UpperBound=Qtr3+1.5*IQR

i_outliers=df[(df['price_per_sqft']<=LowerBound) | (df['price_per_sqft'] >=UpperBound)]
i_non_outliers=df[(df['price_per_sqft'] >=LowerBound) & (df['price_per_sqft'] <=UpperBound)]

print('\nQuarter 1\n', '_'* 25, '\n')
print(Qtr1)
print('\n','-' * 25,'\n')
print('\nQuarter 3\n', '_'* 25, '\n')
print(Qtr3)
print('\n','-' * 25,'\n')
print('\nInter Quartile Range\n', '_'* 25, '\n')
print(IQR)
print('\n','-' * 25,'\n')
print('\nLower Bound\n', '_'* 25, '\n')
print(LowerBound)
print('\n','-' * 25,'\n')
print('\nUpper Bound\n', '_'* 25, '\n')
print(UpperBound)
print('\n','-' * 25,'\n')
print('\nOUTLIERS\n', '_'* 25, '\n')
print(i_outliers)
print('\n','-' * 25,'\n')
print('\nNON OUTLIERS\n' + '_' * 25)
print(i_non_outliers)
```

### d) Z Score method

```python
from scipy.stats import zscore
Z_Score=zscore(df['price_per_sqft'])

Z_outliers=df[(Z_Score <-3 ) | (Z_Score >3)]
Z_non_outliers=df[(Z_Score >-3 ) & (Z_Score <3)]

print('\nZScore \n', '_'* 25, '\n')
print(Z_Score)
print('\n','-' * 25,'\n')
print('\nOUTLIERS\n', '_'* 25, '\n')
print(Z_outliers)
print('\n','-' * 25,'\n')
print('\nNON OUTLIERS\n' + '_' * 25)
print(Z_non_outliers)
```

## ----------------------------------------------------------------------------------------------------

## Creating a box plot and use this to determine which method seems to work best to remove outliers for this data

```python
import matplotlib.pyplot as plt
import seaborn as sns
DataToPlot = {
    'Processed Data': df['price_per_sqft'],
    'Mean and Standard Deviation': non_outliers['price_per_sqft'],
    'Percentile': p_non_outliers['price_per_sqft'],
    'Inter Quartile Range': i_non_outliers['price_per_sqft'],
    'ZScore': Z_non_outliers['price_per_sqft']
}

for method, data in DataToPlot.items():
    plt.figure(figsize=(6, 4))
    sns.boxplot(y=data)
    plt.title(f'Box Plot of Price per sqft ({method})')
    plt.ylabel('Price per sqft')
    plt.tight_layout()
    plt.show()
```

### IN SINGLE PLOT

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

CombinedData = pd.DataFrame({
    'Processed Data': df['price_per_sqft'],
    'Mean and Standard Deviation': non_outliers['price_per_sqft'],
    'Percentile': p_non_outliers['price_per_sqft'],
    'Inter Quartile Range': i_non_outliers['price_per_sqft'],
    'ZScore': Z_non_outliers['price_per_sqft']
})

MeltedData = CombinedData.melt(var_name='Method', value_name='Price per sqft')

plt.figure(figsize=(10, 6))
sns.boxplot(x='Method', y='Price per sqft', data=MeltedData)
plt.title('Comparison of Outlier Removal Methods on Price per sqft')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

### The Inter Quartile Range and Percentile methods are generally better and more reliable choices for identifying outliers compared to the Z-score and Mean and Standard Deviation method.

## Drawing histplot to check the normality of the column(price per sqft column) and perform transformations if needed. Checking the skewness and kurtosis before and after the transformation.

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis
sns.histplot(df['price_per_sqft'],kde=True)
plt.title("Distribution")
plt.xlabel("Price per sqft")      
plt.ylabel("Count")
plt.show()

print("Skewness:", skew(df['price_per_sqft']))
print("Kurtosis:", kurtosis(df['price_per_sqft']))
```

### The price_per_sqft data has a strong right skew (skewness = 108.27) and very high kurtosis (12090.63), meaning most prices are clustered at the lower end, while a few very high values stretch the distribution to the right.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew,kurtosis

df['price_per_sqft_log'] = np.log(df['price_per_sqft'])

sns.histplot(df['price_per_sqft_log'], kde=True)
plt.title("Transformed Distribution")
plt.xlabel("Log(Price per sqft)")      
plt.ylabel("Count")
plt.show()

print("Skewness (log):", skew(df['price_per_sqft_log']))
print("Kurtosis (log):", kurtosis(df['price_per_sqft_log']))
```

### A log transformation was applied to adjust the scale and make the skewed data more normal. As a result, skewness and kurtosis were reduced.
